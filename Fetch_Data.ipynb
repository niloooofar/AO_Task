```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel('data/data.xlsx', header=1)
df.head()
```

```python
sale_df = df.iloc[:,:7]
# sale_df
```

```python
sale_profit_df = df.iloc[:5,9:11]
sale_profit_df.rename(columns={'Product.1': 'Product'}, inplace=True)
# sale_profit_df
```

```python
employee_df = df.iloc[:25, 13:17]
employee_df.columns = ['Id', 'Name', 'Manager', 'ManagerId']
# employee_df
```

```python
# put datas of dataframes into tables in a sql database

conn = sqlite3.connect('data/ahan_online.db')

with conn:
    Sale_df.to_sql('sale', conn, if_exists='replace', index = False)
    sale_profit_df.to_sql('sale_profit', conn, if_exists='replace', index = False)
    employee_df.to_sql('employee', conn, if_exists='replace', index = False);
```
